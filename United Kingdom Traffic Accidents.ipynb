```python
data2 = pd.read_csv('accidents_2009_to_2011.csv')
data2.head()
```

```python
data2.info()
```

```python
data = pd.concat([data1,data2], axis=0)
data.head()
```

```python
data.describe().T
```

```python
data.isnull().values.any()
```

```python
data.isnull().sum()
```

```python
data.drop(["Accident_Index","LSOA_of_Accident_Location"], axis=1, inplace=True)
```

```python
data['Date_time'] =  data['Date'] + ' ' + data['Time']
data['Date_time'] = pd.to_datetime(data.Date_time)
```

```python
data.drop(['Date','Time'],axis =1 , inplace=True)
```

```python
plt.figure(figsize=(14,8))
data.Date_time.dt.dayofweek.hist(bins=7,rwidth=0.6,alpha=0.9, color= 'green')
plt.title('Accidents On The Day Of A Week' , fontsize= 35)
plt.grid(False)
plt.ylabel('Accident Count' , fontsize = 25)
plt.xlabel('0 - Sunday ,  1 - Monday  ,2 - Tuesday , 3 - Wednesday , 4 - Thursday , 5 - Friday , 6 - Saturday' , fontsize = 13)
```

```python
plt.figure(figsize=(12,6))
data.Date_time.dt.hour.hist(rwidth=0.8,alpha =0.8, color= 'blue')
plt.title('Time of the day/night',fontsize= 30)
plt.grid(False)
plt.xlabel('Time 0-24 hours' , fontsize = 20)
plt.ylabel('Accident count' , fontsize = 15)
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data['Day_of_Week'],hue=data['Accident_Severity'])
plt.title('Number of accidents each day',)
plt.legend(title="Accident Severity",labels=["Fatal","Serious","Slight"])
data['Day_of_Week'].value_counts()
plt.ylabel('Accident count' , fontsize = 20)
```

```python
plt.figure(figsize=(14,6))
sns.countplot(data['Speed_limit'],hue=data['Accident_Severity'])
plt.title('The severity of the accident according to the speed limit',size=20)
plt.legend(title="Accident Severity",labels=["Fatal","Serious","Slight"])
plt.ylabel('Accident count' , fontsize = 20)
```

```python
plt.figure(figsize=(14,6))
sns.countplot(data['Road_Surface_Conditions'],hue=data['Accident_Severity'])
plt.title('Relation of accidents with the conditions of the road surface',size=20)
plt.legend(title="Accident Severity",labels=["Fatal","Serious","Slight"])
plt.ylabel('Accident count' , fontsize = 20)
```

```python
hours = pd.DataFrame(data["Date_time"].astype("datetime64"))
hours = pd.DataFrame(hours.Date_time.dt.strftime('%H').sort_values(),columns=["Date_time"])
```

```python
plt.figure(figsize=(22,10))
sns.countplot(hours['Date_time'])
plt.title("Number Of Accidents by Hours")
plt.xlabel("Hours")
plt.ylabel("Number Of Accidents")
```
